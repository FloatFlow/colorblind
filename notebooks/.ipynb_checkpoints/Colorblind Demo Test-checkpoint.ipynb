```python
import numpy as np
import cv2
from colorblind import colorblind
import matplotlib.pyplot as plt
```

# Red-Green Colorblindness

```python
img = cv2.imread('../images/seven.jpg')
img = img[..., ::-1]
plt.imshow(img)
plt.show()
```

## Protanopia

```python
pro_img = colorblind.simulate_colorblindness(img, colorblind_type='protanopia')
cv2.imwrite('../images/protanopia_img.jpg', pro_img[..., ::-1])
plt.imshow(pro_img.astype(np.uint8))
plt.show()
```

### Daltonization Correction

```python
pro_daltonize_img = colorblind.daltonize_correct(img, colorblind_type='protanopia')
cv2.imwrite('../images/protanopia_daltonized_img.jpg', pro_daltonize_img[..., ::-1])
plt.imshow(pro_daltonize_img.astype(np.uint8))
plt.show()
```

```python
pro_daltonize_viewed_img = colorblind.simulate_colorblindness(pro_daltonize_img, colorblind_type='protanopia')
cv2.imwrite('../images/protanopia_daltonized_view_img.jpg', pro_daltonize_viewed_img[..., ::-1])
plt.imshow(pro_daltonize_viewed_img.astype(np.uint8))
plt.show()
```

### Color Blind Filter Service Correction

```python
pro_hsl_img = colorblind.cbfs_correct(img)
cv2.imwrite('../images/protanopia_hsl_img.jpg', pro_hsl_img[..., ::-1])
plt.imshow(pro_hsl_img)
plt.show()
```

```python
pro_hsl_viewed_img = colorblind.simulate_colorblindness(pro_hsl_img, colorblind_type='protanopia')
cv2.imwrite('../images/protanopia_hsl_view_img.jpg', pro_hsl_viewed_img[..., ::-1])
plt.imshow(pro_hsl_viewed_img.astype(np.uint8))
plt.show()
```

### HSV Hue Shift Correction

```python
pro_hsv_img = colorblind.hsv_color_correct(img, colorblind_type='protanopia')
print(np.min(pro_hsv_img), np.max(pro_hsv_img))
cv2.imwrite('../images/protanopia_hsv_img.jpg', pro_hsv_img[..., ::-1])
plt.imshow(pro_hsv_img.astype(np.uint8))
plt.show()
```

```python
pro_hsv_viewed_img = colorblind.simulate_colorblindness(pro_hsv_img, colorblind_type='protanopia')
cv2.imwrite('../images/protanopia_hsv_view_img.jpg', pro_hsv_viewed_img[..., ::-1])
plt.imshow(pro_hsv_viewed_img.astype(np.uint8))
plt.show()
```

### LAB Color Shift

```python
pro_lab_img = colorblind.lab_correct(img, l_shift=30, a_shift=30, b_shift=30)
print(np.min(pro_lab_img), np.max(pro_lab_img))
cv2.imwrite('../images/protanopia_lab_img.jpg', pro_lab_img[..., ::-1])
plt.imshow(pro_lab_img.astype(np.uint8))
plt.show()
```

```python
pro_lab_viewed_img = colorblind.simulate_colorblindness(pro_lab_img, colorblind_type='protanopia')
cv2.imwrite('../images/protanopia_lab_view_img.jpg', pro_lab_viewed_img[..., ::-1])
plt.imshow(pro_lab_viewed_img.astype(np.uint8))
plt.show()
```
